<h2>About dataset</h2>

<h3>Description</h3>

This dataset contains statistics on the world's billionaires, including information about their businesses, industries, and personal details. It provides insights into the wealth distribution, business sectors, and demographics of billionaires worldwide.

<h3>Key Features</h3>

- **rank**: The ranking of the billionaire in terms of wealth.
- **finalWorth**: The final net worth of the billionaire in U.S. dollars.
- **category**: The category or industry in which the billionaire's business operates.
- **personName**: The full name of the billionaire.
- **age**: The age of the billionaire.
- **country**: The country in which the billionaire resides.
- **city**: The city in which the billionaire resides.
- **source**: The source of the billionaire's wealth.
- **industries**: The industries associated with the billionaire's business interests.
- **countryOfCitizenship**: The country of citizenship of the billionaire.
- **organization**: The name of the organization or company associated with the billionaire.
- **selfMade**: Indicates whether the billionaire is self-made (True/False).
- **status**: "D" represents self-made billionaires (Founders/Entrepreneurs) and "U" indicates inherited or unearned wealth.
- **gender**: The gender of the billionaire.
- **birthDate**: The birthdate of the billionaire.
- **lastName**: The last name of the billionaire.
- **firstName**: The first name of the billionaire.
- **title**: The title or honorific of the billionaire.
- **date**: The date of data collection.
- **state**: The state in which the billionaire resides.
- **residenceStateRegion**: The region or state of residence of the billionaire.
- **birthYear**: The birth year of the billionaire.
- **birthMonth**: The birth month of the billionaire.
- **birthDay**: The birth day of the billionaire.
- **cpi_country**: Consumer Price Index (CPI) for the billionaire's country.
- **cpi_change_country**: CPI change for the billionaire's country.
- **gdp_country**: Gross Domestic Product (GDP) for the billionaire's country.
- **gross_tertiary_education_enrollment**: Enrollment in tertiary education in the billionaire's country.
- **gross_primary_education_enrollment_country**: Enrollment in primary education in the billionaire's country.
- **life_expectancy_country**: Life expectancy in the billionaire's country.
- **tax_revenue_country_country**: Tax revenue in the billionaire's country.
- **total_tax_rate_country**: Total tax rate in the billionaire's country.
- **population_country**: Population of the billionaire's country.
- **latitude_country**: Latitude coordinate of the billionaire's country.
- **longitude_country**: Longitude coordinate of the billionaire's country.

<h3>Potential Use Cases</h3>

- **Wealth distribution analysis**: Explore the distribution of billionaires' wealth across different industries, countries, and regions.
- **Demographic analysis**: Investigate the age, gender, and birthplace demographics of billionaires.
- **Self-made vs. inherited wealth**: Analyze the proportion of self-made billionaires and those who inherited their wealth.
- **Economic indicators**: Study correlations between billionaire wealth and economic indicators such as GDP, CPI, and tax rates.
- **Geospatial analysis**: Visualize the geographical distribution of billionaires and their wealth on a map.
- **Trends over time**: Track changes in billionaire demographics and wealth over the years.

1.7 Convert `gdp_country` datatype and change `selfMade` value

- Cột `gdp_country` loại bỏ $ và chuyển sang kiểu dữ liệu dạng numeric.
- Cột `selfMade` sẽ đổi giá trị **True = selfMade**, **False = Inherited** để làm rõ các giá trị của cột này.
- Cột `birthDay` chuyển sang kiểu dữ liệu datetime.

```python
df['gdp_country'] = df['gdp_country'].replace('[$,]', '', regex = True)
df['gdp_country'] = pd.to_numeric(df['gdp_country'], errors = 'coerce')
df['selfMade'] = df['selfMade'].astype(str).replace({'True':'Self-Made', 'False':'Inherited'})
df['birthDate'] = pd.to_datetime(df['birthDate'])
```

1.8 Replacing Null value in `country` and `city`

```python
def replace_missing(df, column, value):
    df[column].fillna(value, inplace=True)
    return df
```

```python
df = replace_missing(df, 'country', 'other')
df = replace_missing(df, 'city', 'other')
```

- Để thuận tiện cho việc phân tích ta sẽ tiến hành thay các giá trị Null của 2 cột `country` và `city` bằng *'other'*

1.9 Fill the Null with mean values

```python
columns_to_fill = ['age', 'gdp_country',
                   'life_expectancy_country','tax_revenue_country_country','total_tax_rate_country',
                   'latitude_country','longitude_country']
```

```python
for col in columns_to_fill:
    mean_value = df[col].mean()
    df[col] = df[col].fillna(mean_value)
```

- Đối với các cột **'age', 'gdp_country', 'life_expectancy_country', 'tax_revenue_country_country', 'total_tax_rate_country', 'latitude_country', 'longitude_country'** ta sẽ thay Null bằng giá trị mean.

Giờ ta hãy cùng tìm hiểu xem chỉ số kinh tế (GDP, CPI, and tax rates) của một quốc gia liên quan như thế nào đến số lượng tỷ phú?

```python
def outlier_iqr(data):
  q_1, q_3 = np.percentile(data,[25,75])
  iqr = q_3 - q_1
  lower_limit = q_1 - (iqr * 1.5)
  upper_limit = q_3 + (iqr * 1.5)
  return np.where((data > upper_limit) | (data < lower_limit))
```

```python
gdp_billionaires = df.groupby('gdp_country')['rank'].count().reset_index()
cpi_billionaires = df.groupby('cpi_country')['rank'].count().reset_index()
tax_billionaires = df.groupby('total_tax_rate_country')['rank'].count().reset_index()

gdp_billionaires = gdp_billionaires.drop(outlier_iqr(gdp_billionaires['rank'])[0].tolist())
cpi_billionaires = cpi_billionaires.drop(outlier_iqr(cpi_billionaires['rank'])[0].tolist())
tax_billionaires = tax_billionaires.drop(outlier_iqr(tax_billionaires['rank'])[0].tolist())

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.scatterplot(data=gdp_billionaires, x='gdp_country', y='rank', legend=False)
plt.title('GDP and Number of Billionaires', fontsize=16)
plt.xlabel('GDP (in Trillions)', fontsize=12)
plt.ylabel('Number of Billionaires', fontsize=12)
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)

plt.subplot(1, 3, 2)
sns.scatterplot(data=cpi_billionaires, x='cpi_country', y='rank', legend=False)
plt.title(' CPI and Number of Billionaires', fontsize=16)
plt.xlabel('CPI', fontsize=12)
plt.ylabel('', fontsize=12)
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)

plt.subplot(1, 3, 3)
sns.scatterplot(data=tax_billionaires, x='total_tax_rate_country', y='rank', legend=False)
plt.title(' Tax rates and Number of Billionaires', fontsize=16)
plt.xlabel('Tax rates', fontsize=12)
plt.ylabel('', fontsize=12)
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)

plt.show()
```

- Sau khi đã loại bỏ các outlier ta có thể được đa phần các tỷ phú đều tập trung ở các nước có các chỉ số kinh tế GDP thuộc dạng không quá cao, có thể là các nước đang phát triển.
- Số lượng tỷ phú ở các nước có CPI thấp cũng rất cao, vì đây là các nước có nên kinh tế ổn định. Rất thích hợp để đầu tư và phát triển.
- Ngoài ra ở các nước có mức thuế trung bình 20-50% cũng tập trung rất đông tỷ phú. Chính sách thuế thấp có thể tạo điều kiện thuận lợi cho việc tích lũy tài sản và thu nhập cá nhân.


```python
gdp_billionaires = df.groupby('gdp_country')['finalWorth'].mean().reset_index()
cpi_billionaires = df.groupby('cpi_country')['finalWorth'].mean().reset_index()
tax_billionaires = df.groupby('total_tax_rate_country')['finalWorth'].mean().reset_index()

gdp_billionaires = gdp_billionaires.drop(outlier_iqr(gdp_billionaires['finalWorth'])[0].tolist())
cpi_billionaires = cpi_billionaires.drop(outlier_iqr(cpi_billionaires['finalWorth'])[0].tolist())
tax_billionaires = tax_billionaires.drop(outlier_iqr(tax_billionaires['finalWorth'])[0].tolist())

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.scatterplot(data=gdp_billionaires, x='gdp_country', y='finalWorth', legend=False)
plt.title('GDP and Mean of final Worth', fontsize=16)
plt.xlabel('GDP (in Trillions)', fontsize=12)
plt.ylabel('Mean of final Worth', fontsize=12)
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)

plt.subplot(1, 3, 2)
sns.scatterplot(data=cpi_billionaires, x='cpi_country', y='finalWorth', legend=False)
plt.title(' CPI and Mean of final Worth', fontsize=16)
plt.xlabel('CPI', fontsize=12)
plt.ylabel('', fontsize=12)
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)

plt.subplot(1, 3, 3)
sns.scatterplot(data=tax_billionaires, x='total_tax_rate_country', y='finalWorth', legend=False)
plt.title(' Tax rates and Mean of final Worth', fontsize=16)
plt.xlabel('Tax rates', fontsize=12)
plt.ylabel('', fontsize=12)
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)

plt.show()
```

- Đối với giáo dục thì hầu hết các nước tỉ lệ tốt nghiệp cao thường tập trung rất đông tỷ phú. Đây cũng là điều tất yếu vì các tỷ phú đa phần đều đến từ các nước phát triển và bản thân họ cũng được nuôi dưỡng trong một môi trường có nên giáo dục tiên tiến.

```python
def get_zodiac_sign(birth_date):
    if pd.isnull(birth_date) or (birth_date.month == 1 and birth_date.day == 1):
        return None
    birth_date = pd.to_datetime(birth_date)
    zodiac_dates = {
        "Aquarius": [(1, 20), (2, 18)],
        "Pisces": [(2, 19), (3, 20)],
        "Aries": [(3, 21), (4, 19)],
        "Taurus": [(4, 20), (5, 20)],
        "Gemini": [(5, 21), (6, 20)],
        "Cancer": [(6, 21), (7, 22)],
        "Leo": [(7, 23), (8, 22)],
        "Virgo": [(8, 23), (9, 22)],
        "Libra": [(9, 23), (10, 22)],
        "Scorpio": [(10, 23), (11, 21)],
        "Sagittarius": [(11, 22), (12, 21)],
        "Capricorn": [(12, 22), (12, 31)] + [(1, 1), (1, 19)]
    }

    for sign, date_ranges in zodiac_dates.items():
        for start_date, end_date in date_ranges:
            if (birth_date.month == start_date and birth_date.day >= end_date) or \
               (birth_date.month == end_date and birth_date.day <= end_date):
                return sign

# Apply the function to create a new column 'ZodiacSign'
df['ZodiacSign'] = df['birthDate'].apply(get_zodiac_sign)
```

```python
plt.figure(figsize=(30, 10))
zo = df['ZodiacSign'].value_counts()
sns.barplot(x=zo.index, y=zo.values)
plt.title('Distribution of Billionaires by Zodiac')
plt.ylabel('Number of Billionaires')
plt.xlabel('Zodiac')
plt.show()
```
